```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\91730\\Data Analysis & Visualisation\\hr_data.csv')
df.head()
```

```python
sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

Hitorgrams represent the data distribution by
forming bins along the raange of the data and then 
drawing bars to show the number of observations
that fall in each bins.

```python
df.describe()
```

```python
bins=[2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

```python
#kde : Kernel density estimate
sns.distplot(df['time_spend_company'],bins=bins,kde=False)
```

```python
# hist:Histogram
sns.distplot(df['time_spend_company'],bins=bins,hist=False)
```

```python
# rug
sns.distplot(df['time_spend_company'],bins=bins,rug=True)
plt.xticks(bins)
```

```python
# hist_kws : Keyword arguments for histogram
sns.distplot(df['time_spend_company'],bins=bins,hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5})
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5},
             kde_kws={'color':'orange','linewidth':3})
plt.xticks(bins)
```

```python
# color
sns.distplot(df['time_spend_company'],bins=bins,rug=True,color='green')
plt.xticks(bins)
```

```python
# Vertical
sns.distplot(df['time_spend_company'],bins=bins,color='green',vertical=True)
plt.xticks(bins)
```

Heatmaps

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flight_df=sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df=flight_df.pivot(index="month",columns="year",values="passengers")
flight_df.head()
```

```python
plt.figure(figsize=(12,6))
ax=sns.heatmap(flight_df)
print(ax)
```

in strip(legend)-lower the number darker the color and lighter the color higher the number. 

so ,as we move from 1949 to 1960 the color is getting lighter means more and more number of customer is increasing year by year.   


```python
# annot : Annotation
plt.figure(figsize=(12,6))
ax=sns.heatmap(flight_df,annot=True,fmt='d')
print(ax)
```

```python
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d',linecolor='k',linewidths='5')
print(ax)
# linecolor -hexadecimal /w,b,k...etc
```

```python
# cmap
plt.figure(figsize=(14,8))
ax=sns.heatmap(flight_df,annot=True,fmt='d',linecolor='k'
               ,linewidth='5',cmap='Blues')
print(ax)
```

```python
# cbar - removeing it
plt.figure(figsize=(12,6))
sns.heatmap(flight_df,cbar=False)
```

```python
# horizontal cbar
grid_kws={'height_ratios':(.4,.05),"hspace":.4}
f,(ax,cbar_ax)=plt.subplots(2,gridspec_kw=grid_kws)
ax=sns.heatmap(flight_df,cbar_kws={"orientation":"horizontal"},ax=ax,cbar_ax=cbar_ax)
```
