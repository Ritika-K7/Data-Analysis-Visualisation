Dist Plot
Plots with Bar Graph and Kernel Density Estimate.

```python
import seaborn as sns
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\91730\\Data Analysis & Visualisation\\hr_data.csv')
df.head()
```

```python
sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

Hitorgrams represent the data distribution by
forming bins along the raange of the data and then 
drawing bars to show the number of observations
that fall in each bins.

```python
df.describe()
```

```python
bins=[2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

```python
#kde : Kernel density estimate
sns.distplot(df['time_spend_company'],bins=bins,kde=False)
```

```python
# hist:Histogram
sns.distplot(df['time_spend_company'],bins=bins,hist=False)
```

```python
# rug
sns.distplot(df['time_spend_company'],bins=bins,rug=True)
plt.xticks(bins)
```

```python
# hist_kws : Keyword arguments for histogram
sns.distplot(df['time_spend_company'],bins=bins,hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5})
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,hist_kws={'color':'red','edgecolor':'blue','linewidth':3,'alpha':0.5},
             kde_kws={'color':'orange','linewidth':3})
plt.xticks(bins)
```

```python
# color
sns.distplot(df['time_spend_company'],bins=bins,rug=True,color='green')
plt.xticks(bins)
```

```python
# Vertical
sns.distplot(df['time_spend_company'],bins=bins,color='green',vertical=True)
plt.xticks(bins)
```
